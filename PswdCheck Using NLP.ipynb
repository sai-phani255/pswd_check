```python
# We need to check the pswd strength by checking character by character of the pswd 
def word_divide_char(inputs):
    characters = []
    for i in inputs:
        characters.append(i)
    return characters
```

```python
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer


#vectorizer = TfidfVectorizer(lowercase=True)
#

vectorizer = TfidfVectorizer(tokenizer=word_divide_char)
x = vectorizer.fit_transform(x)
```

```python
pickle.dump(vectorizer,open('pswd-vect.pkl','wb'))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression as lr
from sklearn.model_selection import train_test_split as tts
```

```python
x_train,x_test,y_train,y_test  =tts(x,y,test_size=0.20,random_state=42)
```

```python
# By changing in few parameters we can use logistic regression for multi class classification
log_class = lr(penalty='l2',multi_class='ovr')
# We can use l1 or l2 but l2 is most commonly is used as parameter of penalty, used to specify the norm used in the penalization
# ovr means one versus rest if multinomila then it uses cross entropy
```

```python
log_class.fit(x_train,y_train)
```

```python
print(log_class.score(x_test,y_test))
```

```python
## Incase of Multinomial

clf = lr(random_state = 0,multi_class='multinomial',solver='newton-cg')
# we have 2 more solvers but the accuracy with newton-cg was high  

clf.fit(x_train,y_train)
print(clf.score(x_test,y_test))
```

```python
X_predict=np.array(["%@123abcd"])
X_predict=vectorizer.transform(X_predict)
y_pred=log_class.predict(X_predict)
print(y_pred)
```
